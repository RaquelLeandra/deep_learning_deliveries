```python
import pandas as pd
import os

path = '../report_results/tables/'
```

```python
names = ['time_norm','pca', 'window']
dropout_rates = ['0','0.2','0.3', '0.4','0.5']

file_name = '{}_results_accuracy_all_{}.csv'.format(names[0],dropout_rates[0]) 
file_path = os.path.join(path, filename)

table = pd.read_csv(file_path).rename(columns={'Unnamed: 0':'# Neurons'})
print(table.to_latex(index=False))
```

```python
for name in names:
    big_table_as_dict ={}
    for dropout in dropout_rates:
        file_name = '{}_results_accuracy_all_{}.csv'.format(name ,dropout) 
        file_path = os.path.join(path, filename)

        table = pd.read_csv(file_path).rename(columns={'Unnamed: 0':'# Neurons'})
        big_table_as_dict[dropout] = table
    big_table = pd.concat(big_table_as_dict,names=['Dropout', 'Row ID'] )
    print('\\begin{table}[]\n\\centering')
    print(big_table.reset_index('Row ID',drop=True).reset_index().to_latex(index=False))
    print('\\caption{Accuracy summary of the '+name+' experiment with different configurations.}')
    print('\\label{tab:acc_'+name+'}\n\\end{table}\n\n')
```
